```python
import matplotlib.pyplot as plt
```

```python
# cizgi grafiği
x=[1,2,3,4,5]
y=[100,200,300,400,500]
plt.plot(x,y, marker="8",linestyle="dashed", color="red")
plt.grid()
plt.xlabel("günler")
plt.ylabel("satışlar")
plt.show()
```

```python
# çizgi grafikleri
import numpy as np

x=np.linspace(0,10,100)
```

```python
#x değerinin sinüs ve cos . cizgi grafiğini tek grafikte göster
```

```python
y1=np.sin(x)
y2=np.cos(x)
```

```python
#plt.plot() direk kullanılırsa tekil cizgi grafiğiverir
plt.figure(figsize=(8,6))
plt.plot(x,y1, label="sin(x)", color="red")
plt.plot(x,y2, label="cos(x)", color="blue")
plt.legend() # grafiğin bilgi alanını aktif eder
plt.show()
```

bar grafiği


```python
categories=["A","B","C","D","E","F"]
values=[5,10,15,20,25,30]
```

```python
plt.bar(categories,values,color=["red","green","blue"])
plt.xlabel("categories")
plt.ylabel("values")
plt.title("kategori bazlı satış grafiği")
plt.show()
```
