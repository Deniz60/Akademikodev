```python
import nltk
nltk.download("stopwords")
```

```python
# Python, PYTHON, pytHON, PYThon
# python, python:
import string
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps= PorterStemmer()
text= df.iloc[0,0]
text=text.lower()
text= text.translate(str.maketrans('','', string.punctuation))
words= text.split()
words=[word for word in words if not word.isdigit()]
#stem -> kelimelerin köküne iner (koşuyor-> koş)
#remove stopwords -> the,at,is
words= [ ps.stem(word) for word in words if word not in stopwords.words("english")]
" ".join(words)
```

```python
def clean_text(text):
    text=text.lower()
    text= text.translate(str.maketrans('','', string.punctuation))
    words= text.split()
    words=[word for word in words if not word.isdigit()]
    #stem -> kelimelerin köküne iner (koşuyor-> koş)
    #remove stopwords -> the,at,is
    words= [ ps.stem(word) for word in words if word not in stopwords.words("english")]
    return " ".join(words)

```

```python
clean_text("merhaba, bu e-posta 45. python dersi hakkındadır.")
```

```python
df['clean_text']= df['text'].apply(clean_text)
```

```python
df.head()
```

```python
x= df['clean_text']
y= df['label_num']
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer()
X_tfidf=tfidf.fit_transform(df['clean_text'])
X_tfidf
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train,y_test = train_test_split(X_tfidf,y,test_size=0.3, random_state=42)
```

```python
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_pred))
# support -> veri sayısı
# precision -> kesinlik -> söylediğim ne kadar dogru
# recall -> duyarlılık -> ne kadarını yakalamışım
# f1-score -> precision-recallin oranı gibi
```
